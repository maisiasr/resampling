```python
import pandas as pd
import numpy as np
from shapely.geometry import LineString

source_dir = '/content/drive/MyDrive/Turkmenistan/Data/Formula_Depth_D2'
datafile = 'D2_SB4_2_Depth.txt'
outfile = '/content/drive/MyDrive/Turkmenistan/Data/Formula_Depth_D2/D2_SB4_2_Depth_1000.csv'

df = pd.read_csv(f"{source_dir}/{datafile}", sep='\t', index_col=0)

# Remove leading/trailing spaces
df.columns = df.columns.str.strip()

print(df.columns)

xcol = df.columns[0]
ycol = df.columns[1]

line = LineString(zip(df[xcol], df[ycol]))

spacing = 1000

distances = np.arange(0, line.length, spacing)

pts = [line.interpolate(d) for d in distances]

out = pd.DataFrame({
    'X':[p.x for p in pts],
    'Y':[p.y for p in pts]
})

out.to_csv(outfile, index=False)

print(f'Original points: {len(df)}')
print(f'Resampled points: {len(out)}')
```
